```python
import yt_dlp
import os

def baixar_video():
    # Pedir o link do YouTube
    url = input("Digite o link do vídeo do YouTube: ")
    
    print("\nEscolha o formato:")
    print("1 - MP4 (vídeo)")
    print("2 - MP3 (áudio)")
    
    formato = input("Digite sua escolha (1 ou 2): ")
    
    if formato == "1":
        baixar_mp4(url)
    elif formato == "2":
        baixar_mp3(url)
    else:
        print("Opção inválida!")

def baixar_mp4(url):
    print("\nEscolha a qualidade do vídeo:")
    print("1 - Baixa (480p)")
    print("2 - Média (720p)")
    print("3 - Alta (1080p)")
    
    qualidade = input("Digite sua escolha (1, 2 ou 3): ")
    
    opcoes = {
        '1': {
            'format': 'best[height<=480]',
            'outtmpl': '%(title)s_480p.%(ext)s'
        },
        '2': {
            'format': 'best[height<=720]',
            'outtmpl': '%(title)s_720p.%(ext)s'
        },
        '3': {
            'format': 'best[height<=1080]',
            'outtmpl': '%(title)s_1080p.%(ext)s'
        }
    }
    
    if qualidade in opcoes:
        try:
            with yt_dlp.YoutubeDL(opcoes[qualidade]) as ydl:
                print("Baixando vídeo...")
                ydl.download([url])
                print("Download concluído!")
        except Exception as e:
            print(f"Erro ao baixar: {e}")
    else:
        print("Qualidade inválida!")

def baixar_mp3(url):
    print("\nEscolha a qualidade do áudio:")
    print("1 - Baixa (128 kbps)")
    print("2 - Média (192 kbps)")
    print("3 - Alta (320 kbps)")
    
    qualidade = input("Digite sua escolha (1, 2 ou 3): ")
    
    opcoes_base = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
        }],
        'outtmpl': '%(title)s.%(ext)s'
    }
    
    qualidades = {
        '1': '128',
        '2': '192', 
        '3': '320'
    }
    
    if qualidade in qualidades:
        opcoes_base['postprocessors'][0]['preferredquality'] = qualidades[qualidade]
        opcoes_base['outtmpl'] = f'%(title)s_{qualidades[qualidade]}kbps.%(ext)s'
        
        try:
            with yt_dlp.YoutubeDL(opcoes_base) as ydl:
                print("Baixando áudio...")
                ydl.download([url])
                print("Download concluído!")
        except Exception as e:
            print(f"Erro ao baixar: {e}")
    else:
        print("Qualidade inválida!")

if __name__ == "__main__":
    print("=== YouTube Downloader ===")
    baixar_video()
```
